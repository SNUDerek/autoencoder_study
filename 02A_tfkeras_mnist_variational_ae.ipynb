```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from functools import reduce
```

```python
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
# add axis for channel and normalize to 0 ~ 1 (based on training max)
x_test = np.expand_dims(x_test, axis=-1)/np.max(x_train)
x_train = np.expand_dims(x_train, axis=-1)/np.max(x_train)
x_shape = x_train[0].shape
print(np.min(x_train), np.max(x_train))
print(x_train.shape)
print(x_shape)
# # add axis for sparse categorical crossentropy
# y_train = np.expand_dims(y_train, axis=-1)
# y_test = np.expand_dims(y_test, axis=-1)
# print(y_train.shape)
```

```python
trn_dataset = tf.data.Dataset.from_tensor_slices((x_train, x_train))
trn_dataset = trn_dataset.shuffle(buffer_size=128).batch(32, drop_remainder=True)
```

```python
tst_dataset = tf.data.Dataset.from_tensor_slices((x_test, x_test))
tst_dataset = tst_dataset.shuffle(buffer_size=128).batch(32, drop_remainder=True)
```

```python
latent_dims = 2
```

```python
# based on the function in the reference
def getsample(args):
    z_mean, z_logv = args
    batch = tf.keras.backend.shape(z_mean)[0]
    dim = tf.keras.backend.int_shape(z_mean)[1]
    eps = tf.random.normal(shape=(batch, dim))
    return z_mean + tf.math.exp(0.5 * z_logv) * eps
```

```python
img = tf.keras.layers.Input(shape=x_shape, name='enc_input')
x = tf.keras.layers.Conv2D(8, kernel_size=2, activation='relu', strides=2, name='conv2d_a')(img)
x = tf.keras.layers.Dropout(rate=0.2, name='dropout_a')(x)
x = tf.keras.layers.Conv2D(8, kernel_size=4, activation='relu', strides=1, name='conv2d_b')(x)
x = tf.keras.layers.Conv2D(8, kernel_size=4, activation='relu', strides=1, name='conv2d_c')(x)
x = tf.keras.layers.Dropout(rate=0.2, name='dropout_b')(x)
x = tf.keras.layers.Conv2D(8, kernel_size=4, activation='relu', strides=1, name='conv2d_d')(x)
x = tf.keras.layers.Conv2D(4, kernel_size=2, activation='tanh', name='conv2d_e')(x)
e = tf.keras.layers.Flatten(name='bottleneck')(x)
z_mean = tf.keras.layers.Dense(latent_dims, name='z_mean')(e)
z_logv = tf.keras.layers.Dense(latent_dims, name='z_logvar')(e)
z = tf.keras.layers.Lambda(getsample, name='get_sample')([z_mean, z_logv])

encoder = tf.keras.Model(img, [z, z_mean, z_logv], name='encoder')
```

```python
encoder.summary()
```

```python
s = tf.keras.layers.Input(shape=(latent_dims, ), name='z_sample')
x = tf.keras.layers.Dense(64, activation='tanh')(s)
x = tf.keras.layers.Reshape((4, 4, 4))(x)
x = tf.keras.layers.Conv2DTranspose(8, kernel_size=4, dilation_rate=2, activation='tanh', input_shape=(latent_dims, ), name='deconv_1')(x)
x = tf.keras.layers.Conv2DTranspose(8, kernel_size=4, dilation_rate=2, activation='tanh', name='deconv_2')(x)
x = tf.keras.layers.Conv2DTranspose(8, kernel_size=4, dilation_rate=2, activation='tanh', name='deconv_3')(x)
x = tf.keras.layers.Conv2DTranspose(8, kernel_size=4, dilation_rate=2, activation='tanh', name='deconv_4')(x)
x = tf.keras.layers.Conv2D(1, kernel_size=1, activation='sigmoid', name='final_conv')(x)

decoder = tf.keras.Model(s, x, name='decoder')
```

```python
decoder.summary()
```

```python
recons = decoder(encoder(img)[0])
vaemdl = tf.keras.Model(img, recons, name='vae')
```

```python
vaemdl.summary()
```

```python
def vae_loss(img, recons):
    bce = tf.keras.losses.BinaryCrossentropy()
    recon_loss = bce(img, recons)
    recon_loss *= reduce(lambda x, y: x*y, x_shape)
    kl_loss = -0.5 * tf.math.reduce_sum(1 + z_logv - tf.math.square(z_mean) - tf.math.exp(z_logv), axis=-1)
    vae_loss = tf.math.reduce_mean(recon_loss + kl_loss)
    return vae_loss
```

```python
vaemdl.compile(optimizer=tf.keras.optimizers.Adam(),
               loss=vae_loss,
               metrics=[tf.keras.metrics.MeanSquaredError()])
```

```python
vaemdl.fit(trn_dataset, epochs=10, verbose=2)
```

```python
vaemdl.evaluate(tst_dataset)
```

```python
outs = vaemdl.predict(x_test)
outs = outs.squeeze(axis=-1)
```
